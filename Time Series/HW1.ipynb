4) **Find the error correction form for unobserved component of TS in ES with additive damped trend:**

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$

**Unobserved components are (in smoothing form): **
$$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$

$$\hat{y}_{t+1|t} = l_t + \phi b_t \Rightarrow \hat{y}_{t|t-1} = l_{t-1} + \phi b_{t-1}$$
$$e_t=y_t-\hat{y}_{t|t-1}=y_t-l_{t-1}-\phi b_{t-1} \Rightarrow y_t=e_t+l_{t-1}+\phi b_{t-1}$$ 
$$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right) = \alpha (e_t+l_{t-1}+\phi b_{t-1}) + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right) = \alpha e_t + l_{t-1} +\phi b_{t-1}$$ 
<br>
$$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1} = \beta \left(\alpha e_t +\phi b_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1} = \beta \alpha e_t + \phi b_{t-1}$$
